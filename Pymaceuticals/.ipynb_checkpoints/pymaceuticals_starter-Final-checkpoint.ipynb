```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="left")
merged_df

# Display the data table for preview
```

```python
# Checking the number of mice.
num_mice = len((merged_df["Mouse ID"]).unique())
print(num_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html
#merged_df.drop_duplicates(subset=['Mouse ID','Timepoint'])
duplicate_id = merged_df.loc[merged_df.duplicated(subset=['Mouse ID','Timepoint']),'Mouse ID'].unique()
print(duplicate_id)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = merged_df.loc[merged_df["Mouse ID"] == 'g989']
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_data = merged_df[merged_df['Mouse ID'].isin(duplicate_mouse)== False]
clean_data = merged_df.drop_duplicates(subset=['Mouse ID','Timepoint'], keep ='first')
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
num_mice_clean = len((clean_data["Mouse ID"]).unique())
num_mice_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#summary["Mean"] = pd.DataFrame(merged_df.groupby(""))
mean_value = clean_data.groupby("Drug Regimen").mean()['Tumor Volume (mm3)']
median_value = clean_data.groupby("Drug Regimen").median()['Tumor Volume (mm3)']
var_value = clean_data.groupby("Drug Regimen").var()['Tumor Volume (mm3)']
std_value = clean_data.groupby("Drug Regimen").std()['Tumor Volume (mm3)']
sem_value = clean_data.groupby("Drug Regimen").sem()['Tumor Volume (mm3)']

summary = pd.DataFrame({"Mean of Tumor Volume":mean_value,
                        "Median of Tumor Volume":median_value,
                        "Variance of Tumor Volume":var_value,
                        "Standard Deviation of Tumor Volume":std_value,
                        "SEM of Tumor Volume":sem_value
                           
})
summary
# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# df.agg()
summary2 = clean_data.groupby("Drug Regimen").agg({'Tumor Volume (mm3)':["mean","median","var","std","sem"]})
summary2
# This method produces everything in a single groupby function
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#regression.ipynb lesson & correlation.ipynb 
x_values = capavgvol['Weight (g)']
y_values = capavgvol['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume')
weight = merged_df.iloc[:,4]
avgtumorvol =merged_df.iloc[:,6]
correlation = st.pearsonr(weight,avgtumorvol)
print(f"The correlation between weight and average tumor volume is {round(correlation[0],2)}")
plt.show()
```
